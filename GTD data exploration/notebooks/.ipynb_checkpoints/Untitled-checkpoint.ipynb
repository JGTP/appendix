```python
def categorise(row):
    attacktype1 = row['attacktype1_txt']
    suicide = row['suicide']
    weapsubtype1 = row['weapsubtype1_txt']
    claimed = row['claimed']
    targtype1 = row['targtype1_txt']
#     print(attacktype1)
#     print(suicide)
#     print(weapsubtype1)
#     print(claimed)
#     print(targtype1)
    label = dftax[(dftax.attacktype1 == attacktype1) & (dftax.suicide == suicide) & (dftax.weapsubtype1 == weapsubtype1) & (dftax.claimed == claimed) & (dftax.targtype1 == targtype1)]['Fifth level'].values
    if len(label) > 0:
        print(label[0])
        return label[0]

subset['label'] = subset.apply(categorise, axis=1)
subset[subset['label'].notna()]
```
